```python
import graphviz

def code_to_flowchart(code_lines, outfile='flowchart_corrected'):
    # We’ll generate the canonical comparison flow you specified.
    # Shapes: Start/End = oval, Input/Output = parallelogram, Decision = diamond.

    dot = graphviz.Digraph(format='png')
    dot.attr(rankdir='TB')

    node_id = 0
    def new_node(label, shape='rectangle'):
        nonlocal node_id
        nid = f"N{node_id}"
        node_id += 1
        dot.node(nid, label, shape=shape)
        return nid

    # Start
    start = new_node('Start', 'oval')

    # Inputs (parallelogram)
    input_a = new_node("Input: a = int(input('enter the first number:'))", 'parallelogram')
    input_b = new_node("Input: b = int(input('enter the second number:'))", 'parallelogram')
    dot.edge(start, input_a)
    dot.edge(input_a, input_b)

    # First decision: if a > b?
    d_if = new_node("If a > b?", 'diamond')
    dot.edge(input_b, d_if)

    # True branch → print A greater (parallelogram) → End (oval)
    out_a = new_node("Output: print('first number is greater than second number')", 'parallelogram')
    end_a = new_node("End", 'oval')
    dot.edge(d_if, out_a, label='True')
    dot.edge(out_a, end_a)

    # False branch → second decision: elif a < b?
    d_elif = new_node("Elif a < b?", 'diamond')
    dot.edge(d_if, d_elif, label='False')

    # Elif True → print B greater → End
    out_b = new_node("Output: print('second number is greater than first number')", 'parallelogram')
    end_b = new_node("End", 'oval')
    dot.edge(d_elif, out_b, label='True')
    dot.edge(out_b, end_b)

    # Elif False → else → print equal → End
    out_eq = new_node("Output: print('both numbers are equal')", 'parallelogram')
    end_eq = new_node("End", 'oval')
    dot.edge(d_elif, out_eq, label='False')
    dot.edge(out_eq, end_eq)

    dot.render(outfile, cleanup=True)
    print(f"Flowchart saved as {outfile}.png")


# Example usage: mirrors the exact logic requested
code = [
    "a = int(input('enter the first number:'))",
    "b = int(input('enter the second number:'))",
    "if a > b:",
    "    print('first number is greater than second number')",
    "elif a < b:",
    "    print('second number is greater than first number')",
    "else:",
    "    print('both numbers are equal')"
]

if __name__ == '__main__':
    code_to_flowchart(code, outfile='flowchart_corrected')

```

```python
#codes for flowchart with loops
import graphviz

def code_to_flowchart( outfile='flowchart_loops'):
    dot = graphviz.Digraph(format='png')
    dot.attr(rankdir='TB')

    node_id = 0
    def new_node(label, shape='rectangle'):
        nonlocal node_id
        nid = f"N{node_id}"
        node_id += 1
        dot.node(nid, label, shape=shape)
        return nid

    def connect(src, dst, label=None):
        if label:
            dot.edge(src, dst, label=label)
        else:
            dot.edge(src, dst)

    # Start
    start = new_node("Start", "oval")

    # Input: a
    input_a = new_node("Input: a = int(input('enter positive number:'))", "parallelogram")
    connect(start, input_a)

    # While a < 0  (validation loop)
    cond_neg = new_node("While a < 0?", "diamond")
    connect(input_a, cond_neg)

    neg_msg = new_node("Output: print('number is negative!')", "parallelogram")
    input_again = new_node("Input: a = int(input('enter positive number:'))", "parallelogram")
    connect(cond_neg, neg_msg, label="True")
    connect(neg_msg, input_again)
    connect(input_again, cond_neg)  # back to condition (cycle)

    # If not negative, proceed
    fact_init = new_node("Process: fact = 1", "rectangle")
    connect(cond_neg, fact_init, label="False")

    i_init = new_node("Process: i = 1", "rectangle")
    connect(fact_init, i_init)

    # While i <= a  (factorial loop)
    cond_fact = new_node("While i <= a?", "diamond")
    connect(i_init, cond_fact)

    mult = new_node("Process: fact = fact * i", "rectangle")
    inc = new_node("Process: i = i + 1", "rectangle")
    connect(cond_fact, mult, label="True")
    connect(mult, inc)
    connect(inc, cond_fact)  # back to condition (cycle)

    # After factorial loop
    output_fact = new_node("Output: print('the factorial of ', a, ' is = ', fact)", "parallelogram")
    connect(cond_fact, output_fact, label="False")

    # End
    end = new_node("End", "oval")
    connect(output_fact, end)

    dot.render(outfile, cleanup=True)
    print(f"Flowchart saved as {outfile}.png")


if __name__ == '__main__':
    code_to_flowchart(outfile='flowchart_loops')
```
